```python
titanic_df.dropna(subset=['age'], axis=0)
# subset은 특정 열만 선택해서 삭제가능
```

```python
# SimpleImputer
from sklearn.impute import SimpleImputer
imr = SimpleImputer(missing_values=np.nan, strategy='mean')
imr = imr.fit(titanic_df[['age']])
imputed_data = imr.transform(titanic_df[['age']])
imputed_data
# SimpleImputer를 통해 평균값으로 NaN값 채우기, median/ most_frequent도 가능
```

```python
titanic_df[['age']].fillna(method='ffill', axis=0)
# ffill을 통해 바로 앞의 값으로 NaN값을 채울 수 있음
```

```python
titanic_df[['age']].fillna(method='bfill', axis=0)
# bfill을 통해 바로 뒤의 값으로 NaN값을 채울 수 있음
```

```python
import pandas as pd
df = pd.DataFrame([
    ['green', 'M', 10.1, 'class2'],
    ['red', 'L', 13.5, 'class1'],
    ['blue', 'XL', 15.3, 'class2']])
df.columns = ['color', 'size', 'price', 'classlabel']
df
```

```python
size_mapping = {
    'XL':3,
    'L':2,
    'M':1}
```

```python
df['size'] = df['size'].map(size_mapping)
df
```

```python
inv_size_mapping = {v: k for k, v in size_mapping.items()}
df['size'].map(inv_size_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X = df[['color', 'size', 'price']].values
color_le = LabelEncoder()
X[:,0] = color_le.fit_transform(X[:, 0])
X
# blue = 0
# green = 1
# red = 2
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# 1. 가상의 데이터 생성 (실제 데이터 대신 사용)
# 특성(feature) 데이터 X: 100개의 샘플, 각 샘플은 2개의 특성을 가짐
X = np.random.rand(100, 2)
# 타겟(target) 데이터 y: 100개의 샘플에 대한 레이블 (0 또는 1)
y = np.random.randint(0, 2, 100)

print("원본 X의 형태:", X.shape)
print("원본 y의 형태:", y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("\n훈련 세트 X_train의 형태:", X_train.shape)
print("훈련 세트 y_train의 형태:", y_train.shape)
print("테스트 세트 X_test의 형태:", X_test.shape)
print("테스트 세트 y_test의 형태:", y_test.shape)
```
